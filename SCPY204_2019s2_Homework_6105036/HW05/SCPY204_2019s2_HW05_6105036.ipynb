```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#All Data
```

```python
from astroquery.sdss import SDSS
query = "SELECT s.z , s.zErr, p.petroMag_g , p.petroMagErr_g FROM PhotoObjAll AS p JOIN SpecObjAll AS s ON s.bestobjid = p.objid WHERE  s.z BETWEEN 0.3 AND 0.4 AND s.zWarning = 0"
res = SDSS.query_sql(query)
print(res)
```

```python
cols = ['z','petroMag_g']
data = res[cols]
z = data['z']
petroMag_g = data['petroMag_g']
```

```python
plt.plot(z,petroMag_g,'o',color='red')
plt.xlabel("z(Final Redshift)")
plt.ylabel("petroMag_g(Petrosian magnitude)")
plt.grid(True)
```
